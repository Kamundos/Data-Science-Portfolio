### Using Apply on a column
* Increase the value of MPG_City columns by +3

```python
# define a function where you can multiply the column by three
def three_multiple(frame_data): 
  col = frame_data["MPG_City"]
  multiply = col+3
  return multiply
```

The code above works perfectly, however I couldn't get it to actually work on my full apply and needed to use lambda instead

```python
# use lambda as part of the apply function to solve this problem, the lambda is a lot easier to use and saves a lot on the memory of the dataset in general
data["MPG_City"] = data["MPG_City"].apply(lambda x:x+3)
```
